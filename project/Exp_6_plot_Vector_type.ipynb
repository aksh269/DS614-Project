```python
"""
Plot DuckDB Vector Type Experiment Results
------------------------------------------

Reads:
    duckdb_vector_type_results.csv

Plots:
1. Q1 Average Runtime
2. Q6 Average Runtime
3. Database Size

Save this as:
    plot_results.py
"""

import pandas as pd
import matplotlib.pyplot as plt



CSV_FILE = "duckdb_vector_type_results.csv"

df = pd.read_csv(CSV_FILE)

print(df)
```

```python
plt.figure(figsize=(8, 5))

plt.bar(
    df["vector_type"],
    df["q1_avg"]
)

plt.xlabel("Vector Type")
plt.ylabel("Average Runtime (seconds)")
plt.title("Q1 Average Runtime by Vector Type")

plt.tight_layout()

plt.savefig("q1_avg_runtime.png")

plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.bar(
    df["vector_type"],
    df["q6_avg"]
)

plt.xlabel("Vector Type")
plt.ylabel("Average Runtime (seconds)")
plt.title("Q6 Average Runtime by Vector Type")

plt.tight_layout()

plt.savefig("q6_avg_runtime.png")

plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.bar(
    df["vector_type"],
    df["db_size_mb"]
)

plt.xlabel("Vector Type")
plt.ylabel("Database Size (MB)")
plt.title("Database Size by Vector Type")

plt.tight_layout()

plt.savefig("db_size_comparison.png")

plt.show()

print("\n[INFO] Plots saved:")
print(" - q1_avg_runtime.png")
print(" - q6_avg_runtime.png")
print(" - db_size_comparison.png")
```
